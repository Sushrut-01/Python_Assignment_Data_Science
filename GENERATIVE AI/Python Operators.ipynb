```python
#2, Perform @ario$s assignment operations on a @ariable.

# Initial value of the variable
variable = 10

# Addition assignment
variable += 5  # Equivalent to variable = variable + 5
print(f"After addition assignment: {variable}")  # Output: 15

# Subtraction assignment
variable -= 3  # Equivalent to variable = variable - 3
print(f"After subtraction assignment: {variable}")  # Output: 12

# Multiplication assignment
variable *= 2  # Equivalent to variable = variable * 2
print(f"After multiplication assignment: {variable}")  # Output: 24

# Division assignment
variable /= 4  # Equivalent to variable = variable / 4
print(f"After division assignment: {variable}")  # Output: 6.0

# Modulus assignment
variable %= 5  # Equivalent to variable = variable % 5
print(f"After modulus assignment: {variable}")  # Output: 1.0

# Exponentiation assignment
variable **= 3  # Equivalent to variable = variable ** 3
print(f"After exponentiation assignment: {variable}")  # Output: 1.0

# Floor division assignment
variable //= 2  # Equivalent to variable = variable // 2
print(f"After floor division assignment: {variable}")  # Output: 0.0




---------------------------------
In this example, the variable starts at 10, and various arithmetic assignment operations are performed on it:

+= adds 5 to the variable.
-= subtracts 3 from the variable.
*= multiplies the variable by 2.
/= divides the variable by 4.
%= finds the remainder of the variable divided by 5.
**= raises the variable to the power of 3.
//= performs floor division on the variable by 2.
The outputs show the value of the variable after each operation.
```

```python
#3, Compare two n$mbers and print the res$lts.


# Define the two numbers
num1 = 10
num2 = 5

# Compare the numbers
print(f"{num1} == {num2} : {num1 == num2}")  # Equality
print(f"{num1} != {num2} : {num1 != num2}")  # Inequality
print(f"{num1} > {num2} : {num1 > num2}")    # Greater than
print(f"{num1} < {num2} : {num1 < num2}")    # Less than
print(f"{num1} >= {num2} : {num1 >= num2}")  # Greater than or equal to
print(f"{num1} <= {num2} : {num1 <= num2}")  # Less than or equal to
----------------------


In this example, the numbers 10 and 5 are compared using different comparison operators:

== checks if the numbers are equal.
!= checks if the numbers are not equal.
> checks if the first number is greater than the second number.
< checks if the first number is less than the second number.
>= checks if the first number is greater than or equal to the second number.
<= checks if the first number is less than or equal to the second number.
The output will be:
    
    
    10 == 5 : False
10 != 5 : True
10 > 5 : True
10 < 5 : False
10 >= 5 : True
10 <= 5 : False

    
    
    
```

```python
#4, Check conditions $sing logical operators.


# Define two numbers
num1 = 10
num2 = 5

# Define some conditions
condition1 = num1 > num2  # True, because 10 is greater than 5
condition2 = num1 < 20    # True, because 10 is less than 20
condition3 = num2 == 5    # True, because num2 is equal to 5
condition4 = num1 == 15   # False, because num1 is not equal to 15

# Using logical AND operator
result_and = condition1 and condition2  # Both conditions must be True
print(f"condition1 AND condition2: {result_and}")  # Output: True

# Using logical OR operator
result_or = condition1 or condition4  # At least one condition must be True
print(f"condition1 OR condition4: {result_or}")  # Output: True

# Using logical NOT operator
result_not = not condition4  # Inverts the condition
print(f"NOT condition4: {result_not}")  # Output: True

# Combining logical operators
combined_result = (condition1 and condition2) or condition3
print(f"(condition1 AND condition2) OR condition3: {combined_result}")  # Output: True

-------------------------
In this example:

condition1 checks if num1 is greater than num2.
condition2 checks if num1 is less than 20.
condition3 checks if num2 is equal to 5.
condition4 checks if num1 is equal to 15.
Logical operators are then used to combine these conditions:

and operator returns True if both conditions are True.
or operator returns True if at least one condition is True.
not operator inverts the condition.



The outputs show the results of the logical operations:
 
condition1 AND condition2: True
condition1 OR condition4: True
NOT condition4: True
(condition1 AND condition2) OR condition3: True


This demonstrates how 
logical operators can be used 
to check multiple conditions and 
combine their results.
    
    
    

```

```python
#5, Check the identity of @ariables.


# Define two variables
a = [1, 2, 3]
b = a  # b is assigned to refer to the same list as a
c = [1, 2, 3]  # c is a different list with the same contents

# Check the identity of variables
print(f"a is b: {a is b}")  # Output: True, because both a and b refer to the same object
print(f"a is c: {a is c}")  # Output: False, because a and c refer to different objects with the same contents
print(f"a == c: {a == c}")  # Output: True, because the contents of a and c are equal

# Check the identity of integers
x = 10
y = 10
print(f"x is y: {x is y}")  # Output: True, because small integers are cached and reused by Python

# Check the identity of strings
str1 = "hello"
str2 = "hello"
print(f"str1 is str2: {str1 is str2}")  # Output: True, because strings are interned by Python


----------------------------
In this example:

a and b refer to the same list object, so a is b is True.
a and c are different list objects with the same contents, so a is c is False but a == c is True.
x and y are small integers, which are cached and reused by Python, so x is y is True.
str1 and str2 are string literals, which are interned by Python, so str1 is str2 is True.

-------------------------------
a is b: True
a is c: False
a == c: True
x is y: True
str1 is str2: True



This demonstrates how to check the 
identity of variables in Python.

```

```python
#6 Perform bitwise operations on any two integers.


# Define the two integers
num1 = 10  # In binary: 1010
num2 = 5   # In binary: 0101

# Bitwise AND
and_result = num1 & num2
print(f"{num1} & {num2} = {and_result}")  # Output: 0 (0000)

# Bitwise OR
or_result = num1 | num2
print(f"{num1} | {num2} = {or_result}")  # Output: 15 (1111)

# Bitwise XOR
xor_result = num1 ^ num2
print(f"{num1} ^ {num2} = {xor_result}")  # Output: 15 (1111)

# Bitwise NOT
not_result_num1 = ~num1
not_result_num2 = ~num2
print(f"~{num1} = {not_result_num1}")  # Output: -11 (in binary: ...11110101, which is -11 in two's complement)
print(f"~{num2} = {not_result_num2}")  # Output: -6 (in binary: ...11111010, which is -6 in two's complement)

# Bitwise left shift
left_shift_result = num1 << 2
print(f"{num1} << 2 = {left_shift_result}")  # Output: 40 (in binary: 101000)

# Bitwise right shift
right_shift_result = num2 >> 1
print(f"{num2} >> 1 = {right_shift_result}")  # Output: 2 (in binary: 0010)
---------------------------------------

In this example, the integers 10 (binary 1010) and 5 (binary 0101) are used:

Bitwise AND (&): The result is 0 (binary 0000) because there are no positions where both 10 and 5 have 1s.
Bitwise OR (|): The result is 15 (binary 1111) because all positions with at least one 1 result in 1.
Bitwise XOR (^): The result is 15 (binary 1111) because all differing positions result in 1.
Bitwise NOT (~): The result for 10 is -11 and for 5 is -6 because the operation inverts all bits and uses two's complement representation for negative numbers.
Bitwise left shift (<<): Shifting 10 two positions to the left results in 40 (binary 101000).
Bitwise right shift (>>): Shifting 5 one position to the right results in 2 (binary 0010)


--------------------------------------------
The output will be:
    
10 & 5 = 0
10 | 5 = 15
10 ^ 5 = 15
~10 = -11
~5 = -6
10 << 2 = 40
5 >> 1 = 2




```

```python
#7, Use $nary operators to change the sign of a n$mber.

# Define a positive number
positive_number = 10
# Change the sign using unary minus
negative_number = -positive_number
print(f"Negative of {positive_number} is {negative_number}")  # Output: -10

# Define a negative number
negative_number = -20
# Change the sign using unary minus
positive_number = -negative_number
print(f"Positive of {negative_number} is {positive_number}")  # Output: 20

# Using unary plus (optional, does not change the value)
unchanged_number = +positive_number
print(f"Unchanged value of {positive_number} is {unchanged_number}")  # Output: 20
---------------------------
In this example:

The unary minus operator (-) is used to change the sign of the number 10 to -10.
The unary minus operator (-) is used again to change the sign of the number -20 to 20.
The unary plus operator (+) is used to explicitly indicate that the value remains positive, 
but it doesn't change the value.

---------------------------------------------
The output will be:
Negative of 10 is -10
Positive of -20 is 20
Unchanged value of 20 is 20



This demonstrates how to use unary operators to change
the sign of a number in Python.

    


```

```python
#8  Use the ternary operator to assign @al$es based on conditions.

The ternary operator in Python is a one-liner for an if-else statement. It allows you to assign values based on a condition in a concise way.
The syntax is:
    value_if_true if condition else value_if_false

    
Hereâ€™s an example:
    
    # Define two numbers
num1 = 10
num2 = 5

# Use ternary operator to assign a value based on a condition
max_value = num1 if num1 > num2 else num2
print(f"The maximum value is {max_value}")  # Output: 10

# Another example with a different condition
is_even = True
number_type = "Even" if is_even else "Odd"
print(f"The number is {number_type}")  # Output: Even

# Example with a more complex condition
age = 18
status = "Adult" if age >= 18 else "Minor"
print(f"The person is an {status}")  # Output: Adult
-------------------------------------

In these examples:

max_value is assigned the value of num1 if num1 is greater than num2, 
otherwise it is assigned the value of num2.
number_type is assigned "Even" if is_even is True, otherwise it is assigned "Odd".
status is assigned "Adult" if age is greater than or equal to 18, otherwise it is assigned "Minor".

-----------------------------------------
The output will be:
    
    The maximum value is 10
The number is Even
The person is an Adult

    
```
